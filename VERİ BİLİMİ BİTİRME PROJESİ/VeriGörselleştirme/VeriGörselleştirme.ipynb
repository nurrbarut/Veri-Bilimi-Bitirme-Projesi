### 1. R&D harcaması ile kâr arasındaki ilişkiyi scatter plot ile gösterin.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Nur/Documents/50_Startups.csv")

print(df.head())

print(df.columns)

sns.scatterplot(x="R&D Spend", y="Profit", data=df, color="purple", s=100)
plt.title("R&D Harcaması ve Kâr İlişkisi", fontsize=16, color="darkblue", fontweight="bold")
plt.style.use('ggplot')
plt.grid(True, linestyle="--", color="grey", alpha=0.6)
plt.xlabel("R&D Harcaması", fontsize=13, color="green", fontweight="bold")
plt.ylabel("Kâr", fontsize=13, color="green", fontweight="bold")
plt.show()

```

<span style="color:red">Analiz:</span> Grafikte R&D harcaması ile kâr arasında pozitif yönlü güçlü bir ilişki görülmektedir. Ar-Ge yatırımı arttıkça şirketlerin kârı da artmaktadır. Bu durum, Ar-Ge’nin kârlılığı artırmada önemli bir faktör olduğunu göstermektedir.

### 2. Yönetim harcaması ile kâr arasındaki ilişkiyi scatter plot ile gösterin.

```python
sns.scatterplot(
    x="Administration", 
    y="Profit", 
    data=df, 
    color="orange", 
    s=100
)
plt.title("Yönetim Harcaması ve Kâr İlişkisi", fontsize=16, color="darkred")
plt.style.use('ggplot')
plt.grid(True, linestyle="--", color="grey", alpha=0.6)
plt.xlabel("Yönetim Harcaması", fontsize=12, color="green")
plt.ylabel("Kâr", fontsize=12, color="green")
plt.show()
```

<span style="color:red">Analiz:</span> Yönetim harcaması ile kâr arasında belirgin bir ilişki bulunmamaktadır. Veriler düzensiz dağılmış olup, yüksek yönetim harcamaları her zaman yüksek kâr getirmemektedir. Bu da yönetim harcamalarının kâr üzerinde doğrudan etkili olmadığını düşündürmektedir.

### 3. Eyaletlere göre ortalama kârları bar chart ile karşılaştırın.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ortalama_kar = df.groupby("State")["Profit"].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(
    x="State", 
    y="Profit", 
    hue="State",     
    data=ortalama_kar,
    palette="Set2",
    legend=False
)

plt.title("Eyaletlere Göre Ortalama Kâr", fontsize=18, color="darkblue", fontweight="bold")
plt.xlabel("Eyalet", fontsize=14, color="black")
plt.ylabel("Ortalama Kâr", fontsize=14, color="black")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

<span style="color:red">Analiz:</span> Bar chart’ta görüldüğü üzere Florida eyaleti ortalama kâr açısından en yüksek değere sahiptir. Bunu sırasıyla New York ve California takip etmektedir. Bu durum, eyalet bazında farklı iş stratejilerinin ve pazar koşullarının kâr üzerinde etkili olabileceğini göstermektedir. Kârdaki bu farklılık, eyaletler arasındaki yönetim, Ar-Ge ve harcama stratejilerindeki farklılıklardan kaynaklanıyor olabilir.

### 4. R&D, yönetim ve pazarlama harcamalarının dağılımını boxplot ile karşılaştırın.

```python
df_melt = df.melt(id_vars="State", 
                  value_vars=["R&D Spend", "Administration", "Marketing Spend"],
                  var_name="Harcamalar",
                  value_name="Değer")

plt.figure(figsize=(10, 6))
sns.boxplot(x="Harcamalar", y="Değer", hue="Harcamalar", data=df_melt, palette="Set3", dodge=False)

plt.title("R&D, Yönetim ve Pazarlama Harcamalarının Dağılımı", fontsize=18, color="darkred", fontweight="bold")
plt.xlabel("Harcama Türü", fontsize=14)
plt.ylabel("Harcanan Miktar", fontsize=14)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

<span style="color:red">Analiz:</span> Bu kutu grafiği, R&D, Yönetim ve Pazarlama harcamalarının dağılımını göstermektedir. R&D harcamaları düşük ve az değişkenlik gösterirken, Yönetim harcamaları orta seviyededir. Pazarlama harcamaları ise en yüksek ve en geniş aralığa sahiptir. Bu durum, firmaların pazarlama bütçelerinde büyük farklılıklar olduğunu, R&D ve yönetim harcamalarının ise daha sabit seyrettiğini göstermektedir.
